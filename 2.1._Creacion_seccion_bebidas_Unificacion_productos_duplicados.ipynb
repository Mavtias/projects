# Unificamos los nombres de los productos que son iguales pero fueron creados con distinto nombre.

```python
# Agrupar el DataFrame por código de producto y obtener el primer nombre de cada grupo
nombres_productos = factura_final2.groupby("Codigo")["Producto"].first()

# Crear un diccionario que mapee cada código de producto a su nombre correspondiente
codigos_nombres = nombres_productos.to_dict()

# Usar el diccionario para cambiar los nombres de los productos en el DataFrame original
factura_final2["Producto"] = factura_final2["Codigo"].map(codigos_nombres)
```
